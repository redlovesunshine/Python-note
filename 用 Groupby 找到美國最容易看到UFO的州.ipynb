```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("http://bit.ly/uforeports")
```

```python
df.head()
```

```python
df_state = df.groupby("State").count() # groupby 依項目集合值 # count 計算數量
```

```python
df_state
```

```python
df_state.sort_values(by ="Time",ascending=False)
```

```python
df_state.sort_values(by ="Time",ascending=False, inplace=True) #sort 不產生新的 dataframe 方法
                                                               #整個改成sort過後的
```

```python
df_state.head()
```

```python
df_state[:10].Time.plot(kind='bar') #draw長條圖的方法
```
