```python
x = np.linspace(-5,5,500)
y = np.sinc(x)

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.plot(x[y>0], y[y>0], 'o')
# 參數的最後 'o' 表示劃出的點以字串 'o' 來呈現，也可以取代成 'x'。
# 也可以使用 lw=size 來調整話線的寬度 (lw = line weight)。
```
