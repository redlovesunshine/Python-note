```python
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader as web
```

```python
df = web.DataReader("AAPL", "yahoo", start = '2012-9-1', end = '2017-8-31')
```

```python
df.head()
```

```python
P = df['Adj Close']
```

```python
P.head()
```

```python
P.plot()
```

```python
# 計算報酬率
```

```python
P.diff() #和前一期的差別
```

```python
r = P.diff()/P
```

```python
r[-100:].plot()
```

```python
#移動平均
```

```python
P.rolling(window = 20).mean()
```

```python
P.plot()
P.rolling(window = 20).mean().plot() #20天移動平均
P.rolling(window = 60).mean().plot() #60天移動平均
```
